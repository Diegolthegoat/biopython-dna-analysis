```python
from Bio import SeqIO
lista=[]
def full_analysis(sequence):
    seq = sequence
   
    for record in SeqIO.parse(sequence, "fasta"):
        gen = {
            "id": record.id,
            "length": len(record.seq),
            "gc": (record.seq.count('G') + record.seq.count('C')) / len(record.seq) * 100
        }
        lista.append(gen)
    return lista
    
def find_highest_gc(sequence):
    highest_gc=0
    id_of_greater_GC=""
    for x in lista:
        if x["gc"]>=highest_gc:
            highest_gc=x["gc"]
            id_of_greater_GC=x["id"]

    print(f"The biggest Guanine Cytosine percentages has a percentage of {highest_gc}% with the gene responsable for that being {id_of_greater_GC}")



def generate_report(sequence):
    print(f" {'Id':<20}{'length':<15}{'GC':<25}")
    print("-----------------------------------------------------")
    for x in lista:
        print(f"{x['id']:<20}{x['length']:<15}{x['gc']:<25.1f}")






full_analysis("brain_genes.fasta")
find_highest_gc("brain_genes.fasta")
generate_report("brain_genes.fasta")
```
