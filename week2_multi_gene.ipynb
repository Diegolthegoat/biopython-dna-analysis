```python
from Bio import SeqIO

BDNF_record = SeqIO.read("sequence.fasta", "fasta")
LRRK2_record = SeqIO.read("LRKK2.FASTA", "fasta")

def gc_content(seq):
    return (seq.count('G') + seq.count('C')) / len(seq) * 100

def analyze_fasta(archive):
    for record in SeqIO.parse(archive, "fasta"):
        print(f"ID for the file uploaded: {record.id}")
        print(f"GC content of dile uploaded: {gc_content(record.seq):.1f}%")
        print(f"The file uploaded has a length of exactly {len(record)} bases")

analyze_fasta("brain_genes.fasta")
```
