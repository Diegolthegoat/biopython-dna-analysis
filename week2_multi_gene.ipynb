```python
from Bio import SeqIO

def basescount(seq):
    seq = seq.upper()

    counts = {
        'A': seq.count('A'),
        'T': seq.count('T'),
        'G': seq.count('G'),
        'C': seq.count('C')
    }

    return counts


def analyze_fasta(archivo):
    best_gc = 0
    best_id = ""

    for record in SeqIO.parse(archivo, "fasta"):
        counts = basescount(record.seq)

        total = counts['A'] + counts['T'] + counts['G'] + counts['C']
        
        if total == 0:
            continue
        
        gc = (counts['G']+counts['C'])/total* 100

        print("ID:", record.id)
        print("GC:", gc)

        if gc >= best_gc:
            best_gc = gc
            best_id = record.id

    print("\nMajor GC:")
    print(best_id, best_gc)


analyze_fasta("brain_genes.fasta")
```
