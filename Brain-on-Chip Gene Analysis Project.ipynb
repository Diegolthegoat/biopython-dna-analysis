```python
# Brain-on-Chip Gene Analysis Project
# Comparing neurological genes related to Alzheimer's and Parkinson's
# BDNF: neuroprotective factor, declines in Alzheimer's
# LRRK2: major Parkinson's risk gene, 151,289 bases
# APP: Alzheimer's amyloid precursor protein
# Author: Diego
# Date: April 2026

from Bio.Seq import Seq
from Bio import SeqIO

BDNF_record = SeqIO.read("sequence.fasta", "fasta")
LRRK2_record = SeqIO.read("LRKK2.FASTA", "fasta")

APP = "ATGCTGCCCGGTTTGGCACTGCTCCTGCTGGCCGCCTGGACGGCTCGGGCAGTGAATGATGGATG"
BDNF = "ATGACCATCCTTTTCCTTACTATGGTTATTTCATACTTTGGTTGCATGAAGGCTACCCGACATGTCCGAT"
#The next sequence is a file of the full BDNF sequence according to NCBI
REAL_BDNF=BDNF_record.seq 
#This is an EXTREMELY shortened version of the actual LRRK2 gene, one of the genes responsable for parkinsons. Because the gene is 7,584 base pairs long, I chose only to add a few, for the sake of the program"
LRRK2= "ATGGCCTCGGGCAGTTGTCAGGGGTGCGAGGAAGACGAGGAGACTCTGAAGAAGTTGATAGTCAGGCTGA"



def gc_content(seq):
    return (seq.count('G') + seq.count('C')) / len(seq) * 100


def compare_genes(seq1, seq2, name1="SEQ1", name2="SEQ2"):
    len1, len2 = len(seq1), len(seq2)

    if len1 > len2:
        print(f"{name1} has a longer sequence than {name2}")
    elif len2 > len1:
        print(f"{name2} has a longer sequence than {name1}")
    else:
        print("Both sequences have equal length")

    gc1 = gc_content(seq1)
    gc2 = gc_content(seq2)

    if gc1 > gc2:
        print(f"{name1} has higher GC content: {gc1:.1f}% vs {gc2:.1f}%")
    else:
        print(f"{name2} has higher GC content: {gc2:.1f}% vs {gc1:.1f}%")


compare_genes(BDNF, REAL_BDNF, "BDNF", "REAL BDNF")




#DAY 2


print(f"ID: {BDNF_record.id}")
print(f"Longitud real: {len(BDNF_record.seq)} bases")
print(f"GC content: {gc_content(BDNF_record.seq):.1f}%")




print(f"ID for LRRK2 file: {LRRK2_record.id}")
sequence = str(LRRK2_record.seq)
print(f"GC content of LRRK2: {gc_content(LRRK2_record.seq):.1f}%")

print(f"LRRK2 has a length of exactly {len(LRRK2_record)} bases")
```
